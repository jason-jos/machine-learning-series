```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic=pd.read_csv('./train.csv')
tips=pd.read_csv("./tips.csv")
```

## ScatterPlot(Numerical-Numerical )

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'])
```

## Barplot

```python
sns.barplot(titanic,x='Pclass',y='Age')
```

```python
sns.barplot(data=titanic,x='Pclass',y='Fare',hue='Sex')
```

### Boxplot (Numerical and Categorical data)

```python
sns.boxplot(data=titanic,x='Sex',y='Age',hue='Survived')
```

```python
sns.histplot(titanic,x='Age',hue='Survived',element='poly')
```

```python
titanic.groupby(['Pclass']).mean(numeric_only=True)['Age'].plot.pie(autopct="%2f")
```

### HeatMap(Categorical vs Categorical)

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Sex']))
```

### Pairplot

```python
sns.pairplot(tips)
```
