```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.read_csv("./covid_toy.csv")
```

```python
df.isna().sum()
```

```python
df.drop(columns=['has_covid'])
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['has_covid']),df['has_covid'],random_state=111,test_size=0.2)
X_train.shape,X_test.shape
```

```python
tranformer=ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(strategy='mean'),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(drop='first',sparse_output=False),['gender','city'])
    ],remainder='passthrough')
```

```python
X_train_trans=tranformer.fit_transform(X_train)
```

```python
X_test_trans=tranformer.transform(X_test)
```
